```python
df = pd.read_csv('E:\ML\Statewise.csv', nrows=8)
df
```

```python
# drop all row with null value
df.dropna()
```

```python
# drop column(axxis=1) in which null value is present
df.dropna(axis = 1)
```

```python
# drop rows in which null value is present 
df.dropna(how = 'any')
```

```python
# drop rows(by default(axis = 0)) and column(axis = 1) in which null value is present 
df.dropna(how = 'any', axis =1)
```

```python
# drop the rows in which all coloumns have null value
df.dropna(how = 'all')
```

```python
# drop the rows in which there are total 2 not null values in the row
df.dropna(thresh = 2)
```

```python
df.dropna(thresh = 5)
```

```python
# drow the row where null value present in the "Negative" column
df.dropna(subset = ['Negative'])
```
