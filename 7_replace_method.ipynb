```python
import pandas as pd
```

```python
df = pd.read_csv("E:\ML\Statewise.csv", nrows = 8)
df
```

```python
# replace Assam to MP
df.replace(to_replace ='Assam' ,value = 'MP' )
```

```python
df.replace('Goa', 'AP')
```

```python
#convert integer value
df.replace(33.0 , 44.0)
```

```python
#replace more value by one value 
df.replace([2848,74551, 1541],0)
```

```python
#replace more value by particular value 
df.replace([2848,74551, 8117, 39911, 1541, 53575],[111,222,333,444,555,666])
```

```python
# replaceing value of particular column(State) 
df.replace({'State' : 'Delhi'},'none')
```

```python
# replacing all the string value by 0
df.replace('[A-Za-z]', 0 )
```

```python
# replacing all the string value by 0(it always required regex)
df.replace( '[A-Za-z]' ,0 , regex = True)
```

```python
# for particular coloumn
df.replace({'State' : '[A-Za-z]'}, 0 ,regex= True  )
```

```python
df.replace( 'Goa' , method= 'ffill')
```

```python
# to replace String by previous value than we use 'ffill' method
df.replace('Assam', method = 'ffill')
```

```python
# to replace String by next value than we use 'bfill' method
df.replace('Assam', method = 'bfill')
```
