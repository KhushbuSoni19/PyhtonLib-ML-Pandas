```python
import pandas as pd
```

```python
df =pd.read_csv("E:\ML\Results.csv", nrows =2)
df
```

# melt method = convert dataframes short format to long format

```python
pd.melt(df)
```

```python
# all the column according to 'Candidate' column
pd.melt(df, id_vars = ['Candidate'])
```

```python
pd.melt(df, id_vars = ['State'])
```

```python
# print state name in dataframe according to Candidate
pd.melt(df, id_vars = ['Candidate'],value_vars = ['State'])
```

```python
# we can give name to the variables and values
pd.melt(df, id_vars = ['Candidate'],value_vars = ['Party'], var_name = 'Categories',value_name = 'Data')
```
