```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'A':[1,2,3,4], 
                    'B':[10,20,30,40]})

df2 = pd.DataFrame({'C':[5,6,7,8],
                    'D':[50,60,70,80]})
display(df1,df2)
```

```python
df1.join(df2) # joint df1 and df2
```

```python
df2.join(df1) #joint df2 to df1
```

```python
df3 = pd.DataFrame({'A':[1,2,3,4], 
                    'B':[10,20,30,40]},index = ['a','b','c','d'])
df3
```

```python
df2.join(df3)# according ton df1 index 
```

```python
 # according to df2 indexing, cz it is in the right side
df3.join(df2,how  = 'right') 
```

```python
 # according to df3 indexing, cz it is in the left side
df3.join(df2,how  = 'left')
```

```python
df3.join(df2,how  = 'outer')
```

```python
df3.join(df2,how  = 'inner')
```

```python
df4 = pd.DataFrame({'A':[5,6,7,8],'D':[50,60,70,80]})
df4
```

```python
df3.join(df4, rsuffix = '_1' )
```

```python
df3.join(df4, lsuffix = '_1' )
```
