```python
import pandas as pd
```

```python
df = pd.read_csv("E:\ML\Statewise.csv", nrows = 8)
df
```

value : scalar, dict, Series, or DataFrame

```python
# if we give '0' As parameter of fillna function the it will print 'NaN' as '0'
df.fillna(0)
```

```python
# print '2' at the place of 'NaN' use scalar value
df.fillna(2)
```

```python
# for specific colomn use dict
df.fillna({'Date':'None', 'State':'None', 'Negative':0, 'Positive':0})
```

```python
# ffill = fill previous value at the place of 'NaN'
df.fillna(method = 'ffill')
```

```python
# same as ffill
df.fillna(method = 'pad')
```

method : {'backfill', 'bfill', 'pad', 'ffill', None}, default None

```python
#fill backward (next) value at the place of Nan

#method : {'backfill', 'bfill', 'pad', 'ffill', None}, default None

df.fillna(method = 'bfill')
```

axis : {0 or 'index', 1 or 'columns'}

```python
# copy  previous rows(axis = 0) value at the place of NaN in particualr column
df.fillna(method = 'ffill', axis = 0)
```

```python
# copy  previous column(axis = 1) value at the place of NaN 
df.fillna(method = 'ffill', axis = 1)
```

```python
# copy  next column(axis = 1) value at the place of NaN 
df.fillna(method = 'bfill', axis = 1)
```

```python
#it used for to print '0' int the 2 place of Nan
df.fillna(value = 0, limit = 2)
```

```python
df.fillna(method="ffill", limit = 2)
```

```python
df.fillna(0,inplace = True)
```
